```python
# You are using Python
class Person:
    # Type your code here
    def __init__(self,name,birthYear):
        self.name=name
        self.birthYear=birthYear
        
    def display(self):
        print(self.name)
        print(self.birthYear)


class Student(Person):
    # Type your code here
    def __init__(self,name,birthYear,department,percentage):
        super().__init__(name,birthYear)
        self.department=department
        self.percentage=percentage
    
    def display(self):
        super().display()
        print(self.percentage)
        
    def iseligible(self,percentage1):
        if(self.percentage>percentage1):
            return True
        else:
            return False

class Staff(Person):
    # Type your code here
    def __init__(self,name,birthYear,salary):
        super().__init__(name,birthYear)
        self.salary=salary
    def display(self):
        super().display()
        print(self.salary)
#         return super().name+"\n"+super().birthYear+"\n"+self.salary


class TeachingStaff(Staff):
    # Type your code here
    def __init__(self,name,birthYear,salary,subject,result):
        super().__init__(name,birthYear,salary)
        self.salary=salary
        self.subject=subject
        self.result=result
    
    def display(self):
#         return super().name+"\n"+super().birthYear+"\n"+super().salary+"\n"+self.subject+"\n"+self.result
        super().display()
        print(self.subject)
        print(self.result)
    
    def calculateSal(self):
        newSalary=(float) (self.salary+ (self.salary*self.result)/100)
        return newSalary
            

class NonTeachingStaff(Staff):
    # Type your code here
    def __init__(self,name,birthYear,salary,lab,experience):
        super().__init__(name,birthYear,salary)
        self.salary=salary
        self.lab=lab
        self.experience=experience
    
    def display(self):
        super().display()
        print(self.lab)
        print(self.experience)
#         return super().name+"\n"+super().birthYear+"\n"+super().salary+"\n"+self.lab+"\n"+self.experience
    
    def calculateSal(self):
        newSalary=(float) (self.salary+(self.salary*self.experience)/100)
        return newSalary


choice=int(input())
name=input()
birthYear=int(input())
if choice==1:
    department=input()
    attendance=float(input())
    stu=Student(name,birthYear,department,attendance)
    stu.display()
    if(stu.iseligible(75.0)):
        print("Yes")
    else:
        print("No")
elif choice==2:
    subject=input()
    result=float(input())
    salary=float(input())
    ts=TeachingStaff(name,birthYear,salary,subject,result)
    newSal=ts.calculateSal()
    ts.display()
    print(newSal)
elif choice==3:
    lab=input()
    exp=float(input())
    salary=float(input())
    nts=NonTeachingStaff(name,birthYear,salary,lab,exp)
    newSal=nts.calculateSal()
    nts.display()
    print(newSal)
```

```python
class WorkerDetail:
        def __init__(self,emp_code,name,salary):
            self.emp_code=emp_code
            self.name=name
            self.salary=salary
        def display(self):
            print("Employee code:{}".format(self.emp_code))
            print("Employee name:{}".format(self.name))
            print("Employee Basic salary:{}".format(self.salary))
        def displayHRA(self):
            hra=0.6*(self.salary)
            print("HRA:{}".format(hra))
        
class OfficerSal(WorkerDetail):
    def __init__(self,emp_code,name,salary):
        super().__init__(emp_code,name,salary)
        self.salary=salary
    def displayDA(self):
        da=0.98*(self.salary)
        print("DA:{}".format(da))
    
class ManagerSal(OfficerSal):
    def __init__(self,emp_code,name,salary):
        super().__init__(emp_code,name,salary)
        self.salary=salary
    def displayCA(self):
        ca=0.2*(self.salary)
        print("CA:{}".format(ca))

emp_code=input()
name=input()
salary=float(input())

obj=ManagerSal(emp_code,name,salary)
obj.display()

obj.displayHRA()
    
obj.displayDA()

obj.displayCA()
```

```python
class Account:
    def __init__(self,accName,accNo,bankName):
        self.accName=accName
        self.accNo=accNo
        self.bankName=bankName
    def display(self):
        print("Account Holder Name:{}".format(self.accName))
        print("Account Number:{}".format(self.accNo))
        print("Bank Name:{}".format(self.bankName))
    
class AdditionInfo1(Account):
    def __init__(self,accName,accNo,bankName,tinNumber):
        super().__init__(accName,accNo,bankName)
        self.tinNumber=tinNumber
    def display(self):
        super().display()
        print("Tin Number:{}".format(self.tinNumber))
class AdditionInfo2(Account):
    def __init__(self,accName,accNo,bankName,orgName):
        super().__init__(accName,accNo,bankName)
        self.orgName=orgName
    def display(self):
        super().display()
        print("Organisation Name:{}".format(self.orgName))
        
accName_inp=input()
accNo_inp=input()
bankName_inp=input()
tinNumber=input()
orgName=input()
ac=Account(accName_inp,accNo_inp,bankName_inp)
ai1=AdditionInfo1(accName_inp,accNo_inp,bankName_inp,tinNumber)
ai2=AdditionInfo2(accName_inp,accNo_inp,bankName_inp,orgName)
choice=int(input())
if choice==1:
    ac.display()
elif choice==2:
    ai1.display()
elif choice==3:
    ai2.display()
    
    
```

```python
class Patient:
    def __init__(self,patient_id,patient_name,age,gender):
        self.patient_id=patient_id
        self.patient_name=patient_name
        self.age=age
        self.gender=gender
    def displayData1(self):
        print(self.patient_id)
        print(self.patient_name)
        print(self.age)
        print(self.gender)
        

class In_Patient(Patient):
    def __init__(self,patient_id,patient_name,age,gender,room_number,consultation_fee,test_fee,days,room_rent_day):
        super().__init__(patient_id,patient_name,age,gender)
        self.room_number=room_number
        self.consultation_fee=consultation_fee
        self.test_fee=test_fee
        self.days=days
        self.room_rent_day=room_rent_day
    def displayData2(self):
        super().displayData1()
        print(self.room_number)
        print(self.consultation_fee)
        print(self.test_fee)
        print(self.days)
        print(self.room_rent_day)
        
class Bill(In_Patient):
    def __init__(self,patient_id,patient_name,age,gender,room_number,consultation_fee,test_fee,days,room_rent_day):
        super().__init__(patient_id,patient_name,age,gender,room_number,consultation_fee,test_fee,days,room_rent_day)
        self.consultation_fee=consultation_fee
        self.test_fee=test_fee
        self.days=days
        self.room_rent_day=room_rent_day
    def calculateTotalBillAmount(self):
        total_bill=self.consultation_fee+self.test_fee+self.days*self.room_rent_day
        return total_bill
    def displayData3(self):
        super().displayData2()
        
pid=input()
pname=input()
age=input()
gender=input()
room_number=input()
consultation_fee=float(input())
test_fee=float(input())
ND=int(input())
rent_room=float(input())

bill=Bill(pid,pname,age,gender,room_number,consultation_fee,test_fee,ND,rent_room)
tb=bill.calculateTotalBillAmount()
bill.displayData3()
print(tb)

```

```python
# You are using Python
class VISACard:
    # Type your code here
    def __init__(self,holder_name,card_number):
        self.holder_name=holder_name
        self.card_number=card_number
    def display_details(self):
        print(holder_name)
        print(card_number)
    def compute_reward_points(self,purchase_type,amount):
        reward=0.01*amount
        print(reward)
        


class HPVISACard(VISACard):
    # Type your code here
    def __init__(self,holder_name,card_number):
        super().__init__(holder_name,card_number)
    def compute_reward_points(self,purchase_type,amount):
        
        if purchase_type=='Fuel':
            reward=amount*0.01+10.0
        else:
            reward=amount*0.01
        
        print(reward)


##Main Program
card_type = input()
holder_name = input()
card_number = input()


purchase_amount = float(input())
purchase_type = input()

# Type your code here
if card_type=='VISA':
    vcard=VISACard(holder_name,card_number)
    vcard.display_details()
    vcard.compute_reward_points(purchase_type,purchase_amount)
elif card_type=='HPVISA':
    hvcard=HPVISACard(holder_name,card_number)
    hvcard.display_details()
    hvcard.compute_reward_points(purchase_type,purchase_amount)
else:
    print("Invalid Choice")
    
        
```
