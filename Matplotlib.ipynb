```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('weather')
plt.plot(days,max_t,label ='Max')
plt.plot(days,min_t,label = 'Min')
plt.plot(days,avg_t,label = 'Avg')

plt.legend()
```

```python
import numpy as np
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel('revenue(bin)')
plt.title('us Tech Stocks')
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.ylabel('revenue(bin)')
plt.title('us Tech Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label = 'revenue')
plt.bar(xpos+0.2,profit,width = 0.4,label='profit')
plt.legend()
```

```python
plt.yticks(ypos,company)

plt.title('us Tech Stocks')
plt.barh(xpos-0.2,revenue,label = 'revenue')
plt.barh(xpos+0.2,profit,label='profit')
plt.legend()
```

```python
blood_sugar = [115,85,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='green')
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='green',histtype = 'step')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('sugar range')
plt.ylabel('Total no. of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins = [80,100,125,150],rwidth = 0.95,color = ['green','orange'],label=['men','women'])
plt.legend()
```

```python
#orientation
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('sugar range')
plt.ylabel('Total no. of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins = [80,100,125,150],rwidth = 0.95,color = ['green','orange'],label=['men','women'],orientation = 'horizontal')
plt.legend()
```

```python
#Pie chart
```

```python
exp_vals =[1400,600,300,410,250]
exp_labels = ["home rent","food","phone/internet bill","car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels,radius = 1.5)
plt.show()
```

```python
plt.axis('equal')
```

```python
plt.pie(exp_vals,labels=exp_labels,radius = 1.5,autopct = '%0.2f%%',shadow = True )
plt.show()
#autopct is for showing percentage in pie chart
```

```python
plt.pie(exp_vals,labels=exp_labels,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0.1,0,0] )
plt.show()
#explode will make the pie compartment come outward a little.
```

```python
plt.pie(exp_vals,labels=exp_labels,radius = 1.5,autopct = '%0.2f%%',shadow = True,explode = [0,0.1,0.1,0,0],startangle = 45 )
plt.show()
#in plt.show you can pass the argument which will be pass down using excel or csv file.
#if you wanna start your pie chart at an angle, this way you can rotate your pie chart the way you like.
```
