```python
import pandas as pd
df = pd.read_csv("D:\excel_files\weatherdetail.csv",parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace = True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
new_df = df.fillna(method = 'ffill')
new_df
#carry forwarded the previous value(temperature)
#in 1 temperature is 32 amd in 4 the temp drops to 0 which is quite fishy so using the method we can takr previous value so that it can look nice.
#vertically
```

```python
new_df = df.fillna(method = 'bfill')
new_df
#backward  fill 
#vertically
```

```python
new_df = df.fillna(method = 'bfill',axis = "columns")
new_df
#backward fill horizontally , row wise
```

```python
new_df = df.fillna(method = 'ffill',limit=2)
new_df
#limit parameter restrict the number of forware or backward fill
```

```python
new_df = df.interpolate()
new_df

#interpolate will help you guess a better value then forward or backward fill
#since the value of temperature in 1 to 5 is decreasing so it uses interpolation to do a better guess.
```

```python
new_df = df.interpolate(method='time')
new_df
#time method is powerfull when you are guessing the missing values coz it keeps in mind abou the othr factor as well.
```

```python
#drop all n.a values
new_df = df.dropna()
new_df
```

```python
#drop if it has all na values
new_df = df.dropna(how="all")
new_df
```

```python
#keep if it has atleast one valid value
new_df = df.dropna(thresh=1)
new_df
```

```python
#insert values like 3rd jan and 4th jan
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
